```python
# source: https://www.bogotobogo.com/python/python-REST-API-Http-Requests-for-Humans-with-Flask.php

from flask import Flask
from flask import jsonify
from flask import request
from flask_ngrok import run_with_ngrok

app = Flask(__name__)
run_with_ngrok(app)

records = [{'name': 'Joe', 'balance': 100},
           {'name': 'Bill', 'balance': 200},
           {'name': 'Mary', 'balance': 300},
           {'name': 'Mark', 'balance': 400}]

@app.route('/', methods=['GET'])
def hello_world():
    return jsonify({'message': 'Simple ledger webapp is up and running!'})

@app.route('/records', methods=['GET'])
def returnAll():
    return jsonify({'records' : records})

@app.route('/records/<string:name>', methods=['GET'])
def get_record(name):
    for record in records:
        if record['name'] == name:
            return jsonify(record)
    return jsonify({'message': 'Record not found'})

@app.route('/records', methods=['POST'])
def add_record():
    new_record = request.get_json()
    records.append(new_record)
    return jsonify(records)

@app.route('/records/<string:name>', methods=['PUT'])
def update_record(name):
    record_data = request.get_json()
    
    for record in records:
        if record['name'] == name:
            record['name'] = record_data['name']
            record['balance'] = record_data['balance']
            break

    return jsonify(records)

@app.route('/records/<name>', methods=['DELETE'])
def delete_record(name):
    global records
    records = [record for record in records if record['name'] != name]
    return jsonify(records)

@app.route('/records/transaction', methods=['PUT'])
def update_records_transaction():
    data = request.get_json()
    sender_name = data['sender']
    receiver_name = data['receiver']
    sent_amt = data['sent_amt']
    
    for record in records:
        if record['name'] == sender_name:
            record['balance'] -= sent_amt
            break
    
    for record in records:
        if record['name'] == receiver_name:
            record['balance'] += sent_amt
            break
    
    return jsonify(records)

if __name__ == "__main__":
    app.run()
```
