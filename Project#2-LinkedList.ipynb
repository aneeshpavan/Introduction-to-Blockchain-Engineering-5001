The **add_to_head** function adds a new node to the linked list's beginning. It takes a node as input and moves the next node of the new node to the top of the list.


---


The **remove_from_tail** function removes the linked list's final node. It begins by determining whether the list is empty, and if it is, it returns None. If the list only contains one node, it eliminates it and sets the head to None. If the list has more than one node, it begins at the top and works its way down until it reaches the second-to-last node. The second to last node's next node is then set to None, essentially eliminating the last node.


---

The **insert_after** method accepts a node and a value as arguments. It begins at the top of the list and works its way down until it finds a node with the matching value. When it discovers the matching node, it sets the next node of the input node to the matching node's next node and sets the matching node's next node to the input node.
