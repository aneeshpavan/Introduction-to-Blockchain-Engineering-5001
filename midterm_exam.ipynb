Say you have a singly linked list with n nodes. Each node has a value and a pointer to the next node. Calculate the number of nodes you need to visit to search for a specific value:

1. Minimum:
2. Maximum:
3. Average:

Answer 1:

The number of nodes you need to visit to search for a specific value in a singly linked list depends on the position of the value in the list.

1. Minimum:
> In the best-case scenario, the value you are searching for is located in the first node of the list, meaning only one node needs to be visited to find it. Therefore, the minimum number of nodes you need to visit is 1.
2. Maximum:
> In the worst-case scenario, the value is not present in the list and you need to visit all n nodes in the list to confirm its absence. Therefore, the maximum number of nodes you need to visit is n.
3. Average:
> In the average-case scenario, if the value is equally likely to be located in any position within the list, then on average, you would need to visit half of the nodes in the list to find it. Therefore, the average number of nodes you need to visit is n/2 if the node is even and n+1/2 if the node is odd.

You are given a singly linked list code in the next code cell. It will create a linked list with 4 nodes. Add a new method called linked_list.insert_before(self, str_to_find, node_to_insert) and test it by calling linked_list.insert_before("Blockchain", Node("Advanced")) and then print("Current Linked List:", linked_list).

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def __iter__(self):
        node = self.head
        while node is not None:
            yield node
            node = node.next

    def __repr__(self):
        nodes = []
        for node in self:
            nodes.append(node.val)
        return " -> ".join(nodes)

    def add_to_tail(self, node):
        if self.head == None:
            self.head = node
            return
        for current_node in self:
            pass
        current_node.set_next(node)

    def remove_from_head(self):
        if self.head == None:
            return None
        temp = self.head
        self.head = self.head.next
        return temp

    # Method to insert before a string
    def insert_before(self, str_to_find, node_to_insert):
        if self.head is None:
            return None
        if self.head.val == str_to_find:
            node_to_insert.next = self.head
            self.head = node_to_insert
            return
        prev_node = None
        current_node = self.head
        while current_node is not None:
            if current_node.val == str_to_find:
                node_to_insert.next = current_node
                prev_node.next = node_to_insert
                return
            prev_node = current_node
            current_node = current_node.next
        return None


class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

    def set_next(self, node):
        self.next = node

    def __repr__(self):
        return self.val

linked_list = LinkedList()
linked_list.add_to_tail(Node("Intro"))
linked_list.add_to_tail(Node("to"))
linked_list.add_to_tail(Node("Blockchain"))
linked_list.add_to_tail(Node("Engineering"))
print("Current Linked List:", linked_list)

#testing by calling the newly defined method
linked_list.insert_before("Blockchain", Node("Advanced"))
print("Current Linked List:", linked_list)
```

*the insert_before method for a singly linked list. The str_to_find and node_to_insert arguments are required for this technique. In the array, it looks for the element that has a value equivalent to str_to_find. If it locates a node with that value, it updates the nodes' next references to put node_to_insert before that node in the list. The head of the list is updated to refer to node_to_insert if the value of the head node is equal to str_to_find. It gives None if it is unable to locate a node with the value str_to_find.*

A secure hash function must provide both preimage resistance and 2nd preimage resistance. Discuss what thay are and how they differ.

Answer 3:

Preimage resistance and second preimage resistance are two crucial characteristics that guarantee a hash function's security.

Preimage resistance makes it computationally challenging to identify the initial message that generated a given hash value when provided a hash value. This feature protects against an attacker discovering an input message that generates the same hash value as the initial message, ensuring the security of the message.

It is computationally challenging to locate another message that yields the same hash value given an initial message thanks to second preimage resistance. By preventing an attacker from discovering a different message that generates the same hash value as the original message, this feature guarantees the identity of a message.

```python
import hashlib
import base64
import base58

s = "Introduction to Blockchain Engineering Midterm Exam"

# Calculate hashes
sha256hash = hashlib.sha256(s.encode('utf-8')).digest()
md5hash = hashlib.md5(s.encode('utf-8')).digest()
sha512hash = hashlib.sha512(s.encode('utf-8')).digest()

# Convert SHA256 hash to different formats
b_string = "".join(f"{byte:02x}" for byte in sha256hash)
b64string = base64.b64encode(sha256hash).decode('utf-8')
b58string = base58.b58encode(sha256hash).decode('utf-8')

# Print results
print("SHA256 Hash Value: ", sha256hash.hex())
print("MD5 Hex Digest: ", md5hash.hex())
print("SHA512 Hex Digest: ", sha512hash.hex())
print("\n")
print("SHA256 hash in Base256 b-string:", b_string)
print("SHA256 hash in Base64 string:", b64string)
print("SHA256 hash in Base58 string:", b58string)
```

*This code determines the hash values of the string "s" provided in SHA256, MD5, and SHA512. The SHA256 hash number is then converted to hexadecimal, a Base256 b-string, a Base64 string, and a Base58 string, among other forms. Finally, it outputs the SHA256 hash value along with all the other hash values in various forms. This program's objective is to show the various ways that hash numbers can be written.*

```python
pip install ecdsa
```

```python
import ecdsa
import secrets

# Choose a curve parameter (secp256k1 is commonly used for Bitcoin)
curve = ecdsa.SECP256k1

# Generate a random private key
prikey = secrets.randbelow(curve.order)

# Calculate the public key
pubkey = prikey * curve.generator

# Define the message to sign
msg = b"Intro to Blockchain Engineering Midterm Exam"

# Create a SHA256 hash of the message
message_hash = hashlib.sha256(msg).digest()

# Sign the hash using the private key
sig = ecdsa.SigningKey.from_string(prikey.to_bytes(32, byteorder="big"), curve=curve).sign(message_hash)


# Print the private key in hex
print("Random Private Key in hex: ",hex(prikey))
# Print the public key in hex
print("Public Key in hex: ",pubkey)
# Print the signature in hex
print("Signature in hex: ",sig.hex())


vk = ecdsa.VerifyingKey.from_public_point(pubkey, curve=ecdsa.SECP256k1)
# Verify the signature using the public key
try:
    vk.verify(sig, message_hash)
    print("\nTrue")
except ecdsa.BadSignatureError:
    print("\nFalse")
```

*Using the secp256k1 elliptic curve, this code creates a random private key, determines the matching public key, and uses these keys to sign and validate a document.*

*First, a random private key and the elliptic curve are selected. Then, the private key is multiplied by the curve's genesis point to determine the public key.*

*A document is then specified, and its SHA256 hash is generated. The signature produced by signing the message hash with the private key is displayed in hexadecimal notation.*

*The signature is then compared to the message hash using the validating key, which was generated from the public key. The result is "True" if the proof is successful and "False" if it fails.*

```python
# install required packages
!apt-get install libgmp-dev
!pip install bitcoinlib

from pprint import pprint
from bitcoinlib.transactions import *
from bitcoinlib.services.services import Service
```

```python
def validation(addr):
    print("Bitcoin Address:", addr)
    try:
        # Decode base58 address to bytes
        decoded = base58.b58decode(addr)
        # Extract public_key_hash (first 21 bytes) and check_sum (last 4 bytes)
        pubkeyhash = decoded[:-4]
        checksum = decoded[-4:]
        print("Decoded Address (Hex):", pubkeyhash.hex() + checksum.hex())
        # Double SHA256 hash on public_key_hash
        first = hashlib.sha256(pubkeyhash).digest()
        second = hashlib.sha256(first).digest()
        print("Double SHA256 Hash:", second.hex())
        # Verify checksum
        if checksum == second[:4]:
            print("True\n")
        else:
            print("False\n")
    except Exception as e:
        print("Error:", e)
        return False

# Test the function with given addresses
validation("1AL6Wq3cdrvqraCzEpTDGr9ryWGiFCBQsY")
validation("1AL6Wq3cdrvqrbCzEpTDGr9ryWGiFCBQsY")
```

The public key hash is then double-SHA256 hashed, and the first four characters of the resulting hash are compared to the checksum.
```python
# connnect to bitcoin blockchain service provider
srv = Service()

# get a block and pretty print it
b = srv.getblock(400000)
pprint(b.as_dict())
print("\n============================================\n")

# get the next block and pretty print it
b2 = srv.getblock(400001)
pprint(b2.as_dict())
print("\n============================================\n")

# pick a transaction in block #400001
txid = '40e331b67c0fe7750bb3b1943b378bf702dce86124dc12fa5980f975db7ec930'

# get raw (hex) transaction info and print
rt = srv.getrawtransaction(txid)
print("Raw: %s" % rt)

# parse hex tx - this process is called deserialization
t = Transaction.parse_hex(rt)

# create a dictionary of transaction key-value pairs and pprint
pprint(t.as_dict())

print("\n============================================\n")

# verify Tx
print("Verified: %s" % t.verify())
```

```python
from pprint import pprint
from bitcoinlib.transactions import *

# block_hash
block = b.block_hash
print("Block #400000 Block Hash:", block)

# prev_block
prev = b2.prev_block
print("Block #400001 Prev Block:", prev)

# Comparing prev_block and block_hash
if prev == block:
    print("Block #400001 prev_block and Block #400000 block_hash are the same.")
else:
    print("Block #400001 prev_block and Block #400000 block_hash are different.")

# retrieve the coinbase transaction in block #400001
#coinbase transaction is known as the first transaction
cb_transaction = b2.transactions[0]

# get raw (hex) transaction info and print
rt = srv.getrawtransaction(cb_transaction.txid)
print("Raw: %s" % rt)

# parse hex tx - this process is called deserialization
t = Transaction.parse_hex(rt)

# create a dictionary of transaction key-value pairs and pprint
pprint(t.as_dict())

# get the miner's address and mining reward from the coinbase transaction
miner_address = cb_transaction.outputs[0].address
mining_reward = cb_transaction.outputs[0].value

# print the miner's address and mining reward
print("Miner's Address: %s" % miner_address)
print("Mining Reward: %s Satoshis" % mining_reward)
print("\n============================================\n")

# verify Tx
print("Verified: %s" % t.verify())
```
