Problem 3  [15pts]:
Write a python code to print MD5, SHA256 and SHA512 hash values in hex of the the following b-string: b'CpE5001 Intro to Blockchain Engineering at Missouri University of Science and Technology'

````python
import hashlib

string = b'CpE5001 Intro to Blockchain Engineering at Missouri University of Science and Technology'

md5_hash = hashlib.md5(string).hexdigest()
sha256_hash = hashlib.sha256(string).hexdigest()
sha512_hash = hashlib.sha512(string).hexdigest()

print("MD5 Hash:", md5_hash)
print("SHA256 Hash:", sha256_hash)
print("SHA512 Hash:", sha512_hash)
````

The Python code generates MD5, SHA256, and SHA512 hash values of the provided b-string, which is a byte string. The hashlib library is used to perform the hashing. The code converts the byte string to a hash object using the hashlib algorithms and then retrieves the hash value in hex format using the hexdigest() function. Finally, the hash values are printed to the console.

Using this library (choose SHA256 as hash algorithm), write a code to:
1. Build a merkle tree of the following list of b-strings:
[b'Dr.', b'Minsu', b'Choi', b'Blockchain', b'Engineering']
2. Validate membership of b'Choi'
3. Print merkle root hash
4. Insert two additional data items [b'CpE', b'5001']
5. Validate membership of b'CpE'
6. Print merkle root hash again

````python
pip install pymerkle
````

````python
import hashlib
from pymerkle import MerkleTree

# List of data items to be included in the Merkle Tree
data = [b'Dr.', b'Minsu', b'Choi', b'Blockchain', b'Engineering']

# Create a Merkle Tree with SHA256 as hash algorithm
hash_algorithm = hashlib.sha256()
merkle_tree = MerkleTree(data, hash_algorithm)

# Validate membership of 'Choi'
print(merkle_tree.validate(b'Choi'))

# Print Merkle Root Hash
print(merkle_tree.root_hash)

# Insert two additional data items
merkle_tree.insert(b'CpE')
merkle_tree.insert(b'5001')

# Validate membership of 'CpE'
print(merkle_tree.validate(b'CpE'))

# Print Merkle Root Hash again
print(merkle_tree.root_hash)
````

Problem 6 [15pts]:
Write a code to:
1. Create ECDSA private and public key pair. Print the in hex.
2. Generate SHA256 hash of the following b-string:
b'CpE5001 Final exam'
3. Sign the hash using the private key. Print the signature in hex.
4. Use the public key to verify the signature

````python
pip install ecdsa
````

````python
import hashlib
import ecdsa

# Step 1: Create ECDSA private and public key pair. Print the in hex.
sk = ecdsa.SigningKey.generate(curve=ecdsa.SECP256k1)
vk = sk.get_verifying_key()
print("Private key: ", sk.to_string().hex())
print("Public key: ", vk.to_string().hex())

# Step 2: Generate SHA256 hash of the following b-string
data = b'CpE5001 Final exam'
hash_obj = hashlib.sha256(data)
digest = hash_obj.digest()

# Step 3: Sign the hash using the private key. Print the signature in hex.
signature = sk.sign(digest)
print("Signature: ", signature.hex())

# Step 4: Use the public key to verify the signature
verified = vk.verify(signature, digest)
if verified:
    print("Signature verified!")
else:
    print("Invalid signature.")
````
