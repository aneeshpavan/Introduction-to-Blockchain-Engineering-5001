```python
from web3 import Web3, Account
import os

alchemy_url = "https://eth-mainnet.g.alchemy.com/v2/idQ4gmqoEw7p0iLcPu2X4SNl8pbQuc5y"
w3 = Web3(Web3.HTTPProvider(alchemy_url))

# generate 10 random Ethereum wallet addresses and their private keys
for i in range(10):
    # generate a random private key
    private_key = os.urandom(32).hex()
    # derive the corresponding public address
    account = Account.from_key(private_key)
    public_address = account.address
    # print the address and private key
    print(f"Address {i+1}: {public_address}")
    print(f"Private key {i+1}: {private_key}")
    print()
```
