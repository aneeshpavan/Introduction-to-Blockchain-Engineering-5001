```python
import base64
import os
from cryptography.fernet import Fernet
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives.kdf.pbkdf2 import PBKDF2HMAC
from termcolor import colored

password = b"1234ABCD"

ciphertext = b'gAAAAABj99qT2ITPcMEcbp_rJJQPUsACZbCShpEMO0xOt9eVNjx_rm9axn3A8VRr8xFu_2kRnn8GnKuHsvATO77QyWmikdXrla5mhq0qKJGBFUSLAPY4de8qgmqcgahaNV-l9BGqw6WT'

for i in range(256):
    salt = bytes([i])  # try all possible values for a single octet salt
    kdf = PBKDF2HMAC(
        algorithm=hashes.SHA256(),
        length=32,
        salt=salt,
        iterations=480000,
    )
    key = base64.urlsafe_b64encode(kdf.derive(password))
    f = Fernet(key)
    try:
        message = f.decrypt(ciphertext)
        print(f"\033[32mKey: {key.decode()}\033[0m")
        found = key.decode()
    except:
        print(f"Key: {key.decode()}")
```

```python
print(f"Correct Key: \033[32m{found}\033[0m\n")
print(f"Plaintext for the ciphertext is:\n\n \033[33m{message.decode()}\033[0m")
```
