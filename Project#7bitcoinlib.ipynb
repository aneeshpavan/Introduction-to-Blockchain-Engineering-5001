```python
# install required packages
!apt-get install libgmp-dev
!pip install bitcoinlib
from pprint import pprint
from bitcoinlib.transactions import *
from bitcoinlib.services.services import Service
from termcolor import colored
# connnect to bitcoin blockchain service provider
srv = Service()
```

```python
block = srv.getblock(780698)
pprint(block.as_dict())
```

*The above code retrieves information about a block on a blockchain network with a block height of 780698 and prints it out as a dictionary.*

```python
print("All txids in the block in hex:\n")
for transaction in block.transactions:
    pprint({transaction.txid.encode().hex()})
```

*The above code prints out the hexadecimal transaction IDs of all transactions in the previously retrieved block.*

```python
from decimal import Decimal
total_input_satoshis = 0

for transaction in block.transactions:
    for tx_input in transaction.inputs:
        print("The input address {} has a value of {} satoshis".format(tx_input.address, tx_input.value))
        total_input_satoshis += tx_input.value

print(f"\n\033[33mTotal input satoshis: {total_input_satoshis}\033[0m")
total_input_btc = Decimal(total_input_satoshis) * Decimal(10**-8)
print(f"\033[33mTotal input in BTC: {total_input_btc}\033[0m")
```

*The above code iterates through each transaction in the block and each input in each transaction to print the input address and its value in satoshis. It then calculates and prints the total input value of the block in both satoshis and BTC.*

```python
total_output_satoshis = 0

for transaction in block.transactions:
    for tx_output in transaction.outputs:
        print("The output address {} has a value of {} satoshis".format(tx_output.address, tx_output.value))
        total_output_satoshis += tx_output.value

print(f"\n\033[33mTotal output satoshis: {total_output_satoshis}\033[0m")
total_output_btc = Decimal(total_output_satoshis) * Decimal(10**-8)
print(f"\033[33mTotal output in BTC: {total_output_btc}\033[0m")
```

*The above code iterates through each transaction in the block and each output in each transaction to print the output address and its value in satoshis. It then calculates and prints the total output value of the block in both satoshis and BTC.*
