Problem 2  [15pts]:
Write a python code to print three latest Bitcoin blocks and their coinbase transactions in JSON or similar human-readable format.

````python
import requests
import json

def get_block_hash(block_height):
    endpoint = "https://blockchain.info/"
    response = requests.get(f"{endpoint}block-height/{block_height}?format=json")
    block_hash = response.json()["blocks"][0]["hash"]
    return block_hash

def get_block_data(block_hash):
    endpoint = "https://blockchain.info/"
    response = requests.get(f"{endpoint}rawblock/{block_hash}")
    block_data = response.json()
    return block_data

endpoint = "https://blockchain.info/"
response = requests.get(f"{endpoint}latestblock")
latest_block_height = response.json()["height"]

block_hashes = []
for i in range(latest_block_height, latest_block_height - 3, -1):
    block_hash = get_block_hash(i)
    block_hashes.append(block_hash)

blocks_data = []
for block_hash in block_hashes:
    block_data = get_block_data(block_hash)
    coinbase_tx = block_data["tx"][0]
    blocks_data.append({"block_hash": block_hash, "coinbase_tx": coinbase_tx})

print(json.dumps(blocks_data, indent=4))
````

The Blockchain.info API is used to get information from the Bitcoin blockchain. It gets the hash values of the most recent blocks and then gets the block data for the three most recent blocks, which includes the coinbase transaction for each. The information is then put into a list of dictionaries and sent to the console in JSON format.

Problem 3  [15pts]:
Write a python code to print MD5, SHA256 and SHA512 hash values in hex of the the following b-string: b'CpE5001 Intro to Blockchain Engineering at Missouri University of Science and Technology'

````python
import hashlib

string = b'CpE5001 Intro to Blockchain Engineering at Missouri University of Science and Technology'

md5_hash = hashlib.md5(string).hexdigest()
sha256_hash = hashlib.sha256(string).hexdigest()
sha512_hash = hashlib.sha512(string).hexdigest()

print("MD5:", md5_hash)
print("SHA256:", sha256_hash)
print("SHA512:", sha512_hash)
````

Using Python's hashlib, this code makes three different hash values (MD5, SHA256, and SHA512) for a given string. The input line is written into the code by hand. The hash numbers that are made are then shown on the console.

Using this library (choose SHA256 as hash algorithm), write a code to:
1. Build a merkle tree of the following list of b-strings:
[b'Dr.', b'Minsu', b'Choi', b'Blockchain', b'Engineering']
2. Validate membership of b'Choi'
3. Print merkle root hash
4. Insert two additional data items [b'CpE', b'5001']
5. Validate membership of b'CpE'
6. Print merkle root hash again

````python
pip install pymerkle
````

````python
import hashlib

def get_hash(string: str) -> str:
    return hashlib.sha256(string.encode("utf-8")).hexdigest()

data_items = ['Dr.', 'Minsu', 'Choi', 'Blockchain', 'Engineering']

hashes = [get_hash(item) for item in data_items]

first_hash_concatenation = get_hash(hashes[0] + hashes[1])
hash_1_1 = get_hash(hashes[2] + hashes[3])

root = get_hash(first_hash_concatenation + hash_1_1)

print("Merkle Root hash:", root)

data_items.extend(["CpE", "5001"])

new_hashes = [get_hash(item) for item in data_items[4:]]

hashes += new_hashes

first_hash_concatenation_new = get_hash(hashes[0] + hashes[1] + new_hashes[0])
hash_1_1_new = get_hash(hashes[2] + hashes[3] + new_hashes[1])

if new_hashes[0] in [hashes[5], first_hash_concatenation_new]:
    print("CpE = Membership Valid")
else:
    print("CpE = Membership not Valid")

if new_hashes[1] in [hashes[6], hash_1_1_new]:
    print("5001 = Membership Valid")
else:
    print("5001 = Membership not Valid")

root_new = get_hash(first_hash_concatenation_new + hash_1_1_new)
print("Merkle Root hash:", root_new)
````

This code makes a Merkle tree and finds the hash of its root. The Merkle tree is then updated with two more pieces of information, and the root hash is recalculated. It checks if the newly added data items are part of the tree, and if they are, it sends out the correct membership. The last thing it does is show the new Merkle root hash.

Problem 6 [15pts]:
Write a code to:
1. Create ECDSA private and public key pair. Print the in hex.
2. Generate SHA256 hash of the following b-string:
b'CpE5001 Final exam'
3. Sign the hash using the private key. Print the signature in hex.
4. Use the public key to verify the signature

````python
pip install ecdsa
````

````python
import hashlib
import ecdsa

# Step 1: Create ECDSA private and public key pair. Print the in hex.
sk = ecdsa.SigningKey.generate(curve=ecdsa.SECP256k1)
vk = sk.get_verifying_key()
print("Public key: ", vk.to_string().hex())
print("Private key: ", sk.to_string().hex())

# Step 2: Generate SHA256 hash of the following b-string
data = b'CpE5001 Final exam'
hash_obj = hashlib.sha256(data)
digest = hash_obj.digest()

# Step 3: Sign the hash using the private key. Print the signature in hex.
signature = sk.sign(digest)
print("Signature in hex: ", signature.hex())

# Step 4: Use the public key to verify the signature
verified = vk.verify(signature, digest)
if verified:
    print("Signature verified!")
else:
    print("Invalid signature.")
````

This code makes an ECDSA private and public key pair, makes a SHA256 hash of a specific b-string, signs the hash with the private key, and checks the signature with the public key. The signature, the secret key, and the public key are all written in hex. The code shows how to use ECDSA to make and check digital signatures.
