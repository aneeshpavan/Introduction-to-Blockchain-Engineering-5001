```python
# Python 3 code to check 
# available algorithms
  
import hashlib
  
# prints all available algorithms
print ("The available algorithms are : ", end ="")
print (hashlib.algorithms_guaranteed)
```

```python
# initializing string
str = "Aneesh Pavan Prodduturu"
  
# encoding string using encode()
# string is encoded in bytes using UTF-8 (variable length unicode)
# then sending to SHA256
result = hashlib.sha256(str.encode("utf-8"))
 
# printing the equivalent hexadecimal value.
print("The SHA256 hash hexdigest is : ")
print(result.hexdigest())
  
print ("\n")
  
# Change the first letter from A to B - only 1 bit different
str = "Bneesh Pavan Prodduturu"

result = hashlib.sha256(str.encode("utf-8"))
  
print("The SHA256 hash hexdigest is : ")
print(result.hexdigest())
```

The SHA256 hash of two strings: "Aneesh Pavan Prodduturu" and "Bneesh Pavan Prodduturu" is computed using this code. The texts are first encoded in UTF-8 format before being supplied as input to the SHA256 hash function through Python's hashlib package. The generated hash values are given in hexadecimal format. The second string is obtained by altering the first letter of the first string, illustrating that even little changes in the input result in a drastically different hash value.

```python
# initializing string
str = "Missouri S&T"
  
# encoding string using encode()
# string is encoded in bytes using UTF-8 (variable length unicode)
# then sending to MD5
result = hashlib.md5(str.encode("utf-8"))
 
print("The MD5 checksum is : ")
print(result.hexdigest())
```

This code computes the MD5 hash of the text "Missouri S&T" using the hashlib package. The string is initially encoded in UTF-8 format before being sent into the MD5 hash algorithm. The resultant hash value is given in hexadecimal format.

---



The length (# of hex digits) of the MD5 hash generated is 32. The MD5 hash algorithm generates a 128-bit (16-byte) hash value, which is then represented as a 32-digit hexadecimal string.

---



```python
import string

firsttwo = "AB"
target = "cd1d14ba40c6a079487fa6952c7da859"
alphanum = string.digits + string.ascii_letters

for i in alphanum:
    for j in alphanum:
        guessing = firsttwo + i + j
        hash_object = hashlib.md5(guessing.encode())
        hex_digest = hash_object.hexdigest()
        print("Guessed password:", guessing, "with hash:", hex_digest)
        if hex_digest == target:
            print("Found password:", guessing, "with hash:", hex_digest)
            break
    else:
        continue
    break
```

This code is a straightforward brute-force approach that attempts to find a password that yields a specified MD5 hash value. The code creates all feasible combinations of two extra characters from the set of numbers and alphabetic letters when the first two characters of the password, "AB," are fixed. The hashed texts are compared to the target hash value "cd1d14ba40c6a079487fa6952c7da859" using the MD5 function from the hashlib package. If a match is discovered, the code displays the guessed password and its hash value before terminating the search.


---



The brute-force hacking effort was successful because it attempted every possible combination of alphanumeric characters as the password's last two characters until it discovered a match with the target MD5 hash hexdigest. This method can work if the password is short and simple, but it is not secure and should not be used for large-scale or real-world applications.
