```python
import hashlib

# function to get SHA256 hash of a string
def get_hash(string):
    return hashlib.sha256(string.encode("utf-8")).hexdigest()

# data items
data_items = ["Intro to Blockchain Engineering", "Dr. Minsu Choi", "Missouri S&T", "Spring 2023"]

# level 0 hashes
hash_0_1 = get_hash(data_items[0])
hash_0_2 = get_hash(data_items[1])
hash_0_3 = get_hash(data_items[2])
hash_0_4 = get_hash(data_items[3])

# level 1 hashes
hash_1_1 = get_hash(hash_0_1 + hash_0_2)
hash_1_2 = get_hash(hash_0_3 + hash_0_4)

# root hash
root_hash = get_hash(hash_1_1 + hash_1_2)

# print data items and hashes
print("Data#1:", data_items[0])
print("Hash 0-1:", hash_0_1)
print()
print("Data#2:", data_items[1])
print("Hash 0-2:", hash_0_2)
print()
print("Data#3:", data_items[2])
print("Hash 0-3:", hash_0_3)
print()
print("Data#4:", data_items[3])
print("Hash 0-4:", hash_0_4)
print()
print("Hash 1-1:", hash_1_1)
print("Hash 1-2:", hash_1_2)
print()
print("Root hash:", root_hash)
```

```python
# verifying membership of data item "Missouri S&T"
if get_hash(data_items[2]) == hash_0_3:
    print("\nHash 0-3 matches data item 3 ('Missouri S&T')")
else:
    print("\nHash 0-3 does not match data item 3 ('Missouri S&T')")
    
if get_hash(hash_0_3 + hash_0_4) == hash_1_2:
    print("\nHash 1 matches hashes 0-3 and 0-4")
else:
    print("\nHash 1 does not match hashes 0-3 and 0-4")

if get_hash(hash_1_1 + hash_1_2) == root_hash:
    print("\nRoot hash matches hash 1-1 and 1-2")
else:
    print("\nRoot hash does not match hash 1-1 and 1-2")
```
